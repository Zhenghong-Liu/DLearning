# 概率密度的传递

### 概率密度的传递公式

当进行变量变换 $ (x, y) \to (u, v) $ 时，新的随机变量的概率密度 $ p_{U,V}(u, v) $ 可以通过原来的概率密度 $ p_{X,Y}(x, y) $ 和变换的雅可比行列式计算，公式为：

$$
p_{U,V}(u, v) = p_{X,Y}(x, y) \cdot \left| \det(J) \right|,
$$
其中 $ J $ 是变换的雅可比矩阵，定义为：

$$
J = 
\begin{bmatrix}
\frac{\partial x}{\partial u} & \frac{\partial x}{\partial v} \\[4mm]
\frac{\partial y}{\partial u} & \frac{\partial y}{\partial v}
\end{bmatrix}.
$$

那么u,v是新的坐标系，x,y为原来的坐标系。 这个雅可比矩阵是从新坐标系到旧的坐标系的变换。

并且微元之间存在这样的关系：

$$
dxdy = |J|dudv
$$

### 2. 数学推导
从 $ (x, y) \to (u, v) $， 需要满足**变换后的总概率和原概率一致**。这就是为什么雅可比行列式起到关键作用的原因。
$$
P((u, v) \in \text{区域}) = P((x, y) \in \text{对应区域}).
$$

对密度而言，这等价于：
$$
\int_{R_{uv}} p_{U,V}(u, v) \, du \, dv = \int_{R_{xy}} p_{X,Y}(x, y) \, dx \, dy.
$$

利用面积的变换关系 $ dx \, dy = \det(J) \, du \, dv $

由于概率密度一定是正的，所以这里不需要考虑方向，因此使用|det(J)|,也就是对雅可比行列式的值求绝对值，我们得到：

$ dx \, dy = |\det(J)| \, du \, dv $

$$
\int p(u, v) \, du \, dv = \int p(x, y) \, |det(J)| \, du \, dv.
$$

此时：左右两边都是dudv，那么就可以消除了，得到：
$$
p(u, v) =  |det(J)| p(x, y)
$$

#### 雅可比行列式的来源
雅可比行列式 $ \det(J) $ 表示的是变量变换对面积的伸缩效应，定义为：
$$
J = 
\begin{bmatrix}
\frac{\partial x}{\partial u} & \frac{\partial x}{\partial v} \\[4mm]
\frac{\partial y}{\partial u} & \frac{\partial y}{\partial v}
\end{bmatrix}.
$$

其行列式 $ \det(J) $ 的绝对值表示变换后区域的面积（体积）相对原区域的缩放因子。

---

### 3. 直观理解

1. **雅可比行列式的几何意义**：
   - 如果 $|\det(J)| > 1$，表示变换扩大了原始区域面积；
   - 如果 $|\det(J)| < 1$，表示变换缩小了原始区域面积；
   - 如果 $|\det(J)| = 1$，表示变换保持了区域面积不变。

2. **概率密度的作用**：
   - 原始的 $ p_{X,Y}(x, y) $ 是单位面积上的概率；
   - 面积的伸缩（通过雅可比行列式）会影响单位面积上的概率密度，确保总概率不变。

---

### 总结

概率密度的传递公式来源于概率守恒和面积变换的几何关系。雅可比行列式描述了变换后的区域相对于原始区域的拉伸或压缩，因此它直接影响新的概率密度值。


### 具体问题：已知概率密度p(x, y)，求概率密度p(x+y, 2x-y)

我们已知：
$$
u = x + y, \quad v = 2x - y.
$$


已知联合概率密度函数 $ p(x, y) $，想要找到新的随机变量 $ U = X + Y $ 和 $ V = 2X - Y $ 的联合概率密度函数 $ p(U, V) $。

好，既然涉及到随机变量的变换，我记得应该用到**变换密度公式**。

具体来说，如果有一组随机变量的线性变换，我们可以通过雅可比行列式来传递概率密度函数。

首先，我需要表达新的变量 $ U $ 和 $ V $ 如何用原来的 $ X $ 和 $ Y $ 表示：
$$ U = X + Y $$
$$ V = 2X - Y $$

好了，现在我有了 $ X $ 和 $ Y $ 用 $ U $ 和 $ V $ 表示的形式：
$$ X = \frac{U + V}{3} $$
$$ Y = \frac{2U - V}{3} $$

接下来，我需要计算雅可比行列式 $ |J| $。雅可比矩阵 $ J $ 是由 $ X $ 和 $ Y $ 对 $ U $ 和 $ V $ 的偏导数组成的：

$$
J = 
\begin{pmatrix}
\frac{\partial X}{\partial U} & \frac{\partial X}{\partial V} \\[4mm]
\frac{\partial Y}{\partial U} & \frac{\partial Y}{\partial V}
\end{pmatrix}
= 
\begin{pmatrix}
\frac{1}{3} & \frac{1}{3} \\[4mm]
\frac{2}{3} & -\frac{1}{3}
\end{pmatrix}
$$

那么，雅可比行列式就是：
$$ |J| = \left| \frac{1}{3} \times \left(-\frac{1}{3}\right) - \frac{1}{3} \times \frac{2}{3} \right| = \left| -\frac{1}{9} - \frac{2}{9} \right| = \left| -\frac{3}{9} \right| = \frac{1}{3} $$

此时得到微元的关系:
$$
dxdy = \frac{1}{3} dudv
$$

现在，根据变换密度公式，新的联合概率密度函数 $ p(u, v) $ 可以表示为：

$$ \int p_{U,V}(u, v) \, du \, dv = \int p_{X,Y}(x, y) \, \frac{1}{3} \, du \, dv. $$

$$  p(u, v) = \frac{1}{3} \cdot p\left( x, y \right)  $$


### 绘图，感受一下空间的变换

```python
# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Define the original density function p_X,Y(x, y) for an exponential distribution
def p_XY(x, y, mu1=2, mu2=3, sigma1=1, sigma2=1.3, rho=0.3):
    """
    二维高斯分布的概率密度函数 (PDF)

    参数:
    x, y: 目标点坐标
    mu1, mu2: X 和 Y 的均值
    sigma1, sigma2: X 和 Y 的标准差
    rho: X 和 Y 的相关系数 (-1 <= rho <= 1)
    """
    # 常数项
    norm_const = 1 / (2 * np.pi * sigma1 * sigma2 * np.sqrt(1 - rho**2))
    # 指数项
    z = ((x - mu1)**2 / sigma1**2 - 
         2 * rho * (x - mu1) * (y - mu2) / (sigma1 * sigma2) + 
         (y - mu2)**2 / sigma2**2) / (1 - rho**2)
    return norm_const * np.exp(-0.5 * z)

# Define the transformation and inverse transformation
def transform(x, y):
    u = x + y
    v = 2 * x - y
    return u, v

def inverse_transform(u, v):
    x = (u + v) / 3
    y = (2 * u - v) / 3
    return x, y

# Define the transformed density p_U,V(u, v)
def p_UV(u, v):
    x, y = inverse_transform(u, v)
    jacobian_determinant = 1 / 3  # 雅可比行列式的绝对值
    return p_XY(x, y) * jacobian_determinant  #就是上面计算的公式 p(u, v) = 1/3 * p(x, y)

# Generate grid for visualization
grid_size = 100
x_vals = np.linspace(0, 5, grid_size)
y_vals = np.linspace(0, 5, grid_size)
X, Y = np.meshgrid(x_vals, y_vals)
U, V = transform(X, Y)

# Calculate densities
Z_XY = p_XY(X, Y)  # Original density
Z_UV = p_UV(U, V)  # Transformed density

# Plot the original density p_X,Y
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.contourf(X, Y, Z_XY, levels=20, cmap='Reds')
plt.colorbar(label=r"$p_{X,Y}(x, y)$")
plt.title("Original Density $p_{X,Y}(x, y)$", fontsize=14)
plt.xlabel("$x$", fontsize=12)
plt.ylabel("$y$", fontsize=12)
plt.grid(alpha=0.3)

# Plot the transformed density p_U,V
plt.subplot(1, 2, 2)
plt.contourf(U, V, Z_UV, levels=20, cmap='Blues')
plt.colorbar(label=r"$p_{U,V}(u, v)$")
plt.title("Transformed Density $p_{U,V}(u, v)$", fontsize=14)
plt.xlabel("$u$", fontsize=12)
plt.ylabel("$v$", fontsize=12)
plt.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

### 用数值方法证明微元之间的关系

结果表明，对应微元的面积是相等的

```python
from scipy.integrate import dblquad
# 使用数值积分证明微元对应的积分相等
# Define the original density function for numerical integration
def density_XY(x, y):
    return p_XY(x, y)

# Define the transformed density function for numerical integration
def density_UV(u, v):
    return p_UV(u, v)

# Define small integration ranges around the test point (x_test, y_test)
delta = 0.1  # 微元范围
x_test, y_test = 0.1, 0.1  # 原始分布的测试点
u_test, v_test = transform(x_test, y_test)  # 转换后对应的点

# Integration limits for the original distribution
x_min, x_max = x_test - delta, x_test + delta
y_min, y_max = y_test - delta, y_test + delta

# Integration limits for the transformed distribution
u_min, u_max = transform(x_test - delta, x_test + delta)
v_min, v_max = transform(y_test - delta, y_test + delta)

# Compute numerical integration for the original distribution
integral_XY, _ = dblquad(density_XY, x_min, x_max, lambda x: y_min, lambda x: y_max)

# Compute numerical integration for the transformed distribution
integral_UV, _ = dblquad(density_UV, u_min, u_max, lambda u: v_min, lambda u: v_max)

# Compare results
print(f"Numerical integral in original distribution: {integral_XY}")
print(f"Numerical integral in transformed distribution: {integral_UV}")
```
