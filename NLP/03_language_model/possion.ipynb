```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
am = [15000, 2000]
thm = [0, 0]
pm = [160, 320]

def my_funx(x):
    return am[0] * np.sin(2 * np.pi * x / pm[0] + thm[0]) + am[1] * np.sin(2 * np.pi * x / pm[1] + thm[1]) + 2e6 + 25000


x = np.arange(1, 4493)
y = my_funx(x)

plt.figure(figsize=(20, 5))
plt.plot(x, y)
plt.show()
```

```python
# y = np.random.poisson(y)
plt.figure(figsize=(20, 5))
s = slice(1000, 2000)
plt.plot(x[s], y[s])
plt.show()
```

```python
y_noise = y + np.random.normal(0, np.sqrt(y))
plt.figure(figsize=(30, 5))
plt.plot(x, y_noise)
plt.show()
```

```python
y = np.random.poisson(y)
plt.figure(figsize=(30, 5))
plt.plot(x, y)
plt.show()
```

```python
# y = np.random.poisson(y)
plt.figure(figsize=(20, 5))
s = slice(1000, 2000)
plt.plot(x[s], y[s])
plt.show()
```
